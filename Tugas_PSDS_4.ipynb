**PERULANGAN FOR DAN WHILE**

**Looping/Perulangan**

Python menyediakan dua statement untuk melakukan proses perulangan yaitu for dan while. Diantara kedua statement ini, secara umum for lebih banyak digunakan daripada while.

**For Loop**

Perulangan for digunakan untuk melakukan iterasi di dalam sequence seperti list, tuple, string, dan array atau objek yang dapat diiterasi lainnya.

**For Loop Syntax**

for val in sequence:
 loop body


**For Loop Flowchart**

image.png

**example: Python for Loop**

```python
# Program untuk menghitung jumlah semua angka dalam sebuah list
L1 = [2, 4, 5, 7, 9]
jumlah = 0

# iterasi di dalam list
for i in L1:
  jumlah = jumlah + i

print (jumlah)
```

**The range() function**

```python
# Program untuk melakukan iterasi di dalam list melalui indeks
minuman = ["teh", "boba", "kopi", "susu"]
p = len(minuman)

# iterasi di dalam list melalui indeks
for i in range (p):
  print(minuman[i])
```

**for loop with else**

```python
# contoh for loop menggunakan else
for i in range(p):
  print(minuman[i])
else:
  print("elemen sudah habis")
```

**While Loop**

while loop python digunakan untuk melakukan iterasi pada blok kode selama nilai dari test expression adalah true.

**Syntax of while Loop in Pyhton**

while test_Expression:
 Body of while

**While loop flowchart**

image.png

**Example: Pyhton while Loop**

```python
# Program untuk menjumlahkan semua
# bilangan asli hingga suku ke-n
# jum = 1+2+3+...+n

n = 10
jumlah = 0
i = 1

while i<= n:
  jumlah = jumlah + i
  i = i + 1

print (jumlah)
```

**While loop with else**

sama dengan for loop kita juga dapat mengkombinasikan while loop dengan blok else. Bagian else akan dieksekusi jika nilai pada test expression pada while loop bernilai false. perhatikan contoh berikut:

```python
jum = 0
i = 1
while i <= n:
  jum = jum + i
  i = i + 1
else:
  print(jum) # dieksekusi di luar looping
```
